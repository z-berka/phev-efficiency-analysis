```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

DATA_PATH = "../data/JRC_Dataset_PHEV_test_campaign_20240920.xlsx"

df = pd.read_excel(DATA_PATH, sheet_name="Data PHEV")
df.head()
```

```python
print("Number of trips:", len(df))
print("\nColumns:")
print(df.columns)

print("\nBasic statistics:")
print(df.describe())
```

```python
plt.figure()
plt.hist(df["Distance [km]"].dropna(), bins=20)
plt.xlabel("Distance [km]")
plt.ylabel("Frequency")
plt.title("Distribution of Trip Distance")
plt.show()
```

```python
fuel = df["Fuel Consumed [lt]"]
battery = df["Battery energy [kWh]"]

plt.figure()
plt.scatter(fuel, battery)
plt.xlabel("Fuel Consumed [lt]")
plt.ylabel("Battery Energy [kWh]")
plt.title("Fuel vs Battery Energy Usage")
plt.show()
```

```python
plt.figure()
plt.scatter(df["Mean Velocity [km/h]"], df["Fuel Consumed [lt]"])
plt.xlabel("Mean Velocity [km/h]")
plt.ylabel("Fuel Consumed [lt]")
plt.title("Fuel Consumption vs Mean Speed")
plt.show()
```

```python
plt.figure()
plt.scatter(df["Urban Share"], df["Fuel Consumed [lt]"])
plt.xlabel("Urban Share")
plt.ylabel("Fuel Consumed [lt]")
plt.title("Effect of Urban Driving on Fuel Use")
plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,8))
plt.imshow(corr)
plt.colorbar()
plt.title("Correlation Matrix")
plt.show()
```
